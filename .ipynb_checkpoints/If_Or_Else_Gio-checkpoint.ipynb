Número 1°


```python
idade = int(input("Qual a sua idade? "))

if idade >= 12:
    ingresso = input("Está com o ingresso? Responda com (s) para sim e (n)para não: ").lower()
    
    if ingresso == "s":
        print("Entrada liberada")
    else:
        print("Compre o ingresso")
else:
    print("Idade insuficiente")
```

Número 2º



```python
usuario = input("Digite o nome de usuário: ")

if usuario == "Mestre":
    senha = input("Digite a senha: ")
    
    if senha == "1212":
        print("Acesso permitido")
    else:
        print("Senha incorreta")
else:
    print("Usuário inválido")
```

Número 3°


```python
nota_aluno = float(input("Digite a nota do aluno: "))

if nota_aluno >= 7:
    frequencia = float(input("Digite a frequência (0 a 100): "))
    
    if frequencia >= 75:
        print("Aprovado")
    else:
        print("Reprovado por falta")
else:
    print("Reprovado por nota")
```

Número 4°

```python
numero_inteiro = int(input("Digite um número inteiro: "))

if numero_inteiro > 0:
    if numero_inteiro % 2 == 0:
        print("Par positivo")
    else:
        print("Ímpar positivo")
else:
    print("Número não positivo")
```

Número 7°

```python
numero = int(input("Digite um número inteiro: "))
numero
if numero > 0:
    if numero % 2 == 0:
        if numero % 3 == 0:
            print("Múltiplo de 2 e de 3")
        else:
            print("Múltiplo apenas de 2")
    else:
        print("Não é múltiplo de 2")
else:
    print("Número não positivo")
```

Número 8°

```python
nome = input("Digite o seu nome: ")

if nome != "":
    idade = int(input("Digite a sua idade: "))
    
    if idade >= 18:
        print("Cadastro completo")
    else:
        print("Cadastro incompleto: menor de idade")
else:
    print("Nome obrigatório")
```

Número 9°

```python
idade = int(input("Digite a sua idade: "))

if idade >= 60:
    dor_forte = input("Sente dor forte? Responda com (s) para sim e (n)para não: ").lower()
    
    if dor_forte == "s":
        print("Atendimento prioritário imediato")
    else:
        print("Aguardar triagem")
else:
    print("Atendimento comum")
```

Número 10°

```python
usuario = input("Digite o usuário: ")

if usuario == "professor":
    senha = input("Digite a senha: ")
    
    if senha == "prof1234":
        nivel = int(input("Digite o nível de acesso (1 ou outro): "))
        
        if nivel == 1:
            print("Acesso básico")
        else:
            print("Acesso avançado")
    else:
        print("Senha inválida")
else:
    print("Usuário não autorizado")
```
